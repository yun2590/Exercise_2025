```
intList=range(10)
try:
  idx=int(input('숫자를 입력하세요: '))
  print(intList[idx])
except ValueError:
  print('입력한 내용이 숫자가 아닙니다!')
except IndexError as e:
  print(f'인덱스 오류가 발생: {e}')
```
