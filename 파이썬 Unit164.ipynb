```
from shutil import rmtree

folder='/content/myfolder'
k=input(f'<{folder}>를 완전히 삭제하시겠습니까? (y/n): ')
if k.lower()=='y':
  rmtree(folder)
  print(f'<{folder}>를 완전히 삭제했습니다.')
else:
  print(f'<{folder}>를 삭제하지 않습니다.')
```
